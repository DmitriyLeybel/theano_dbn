# Training of wine classification dataset 

```python
from keras.models import Sequential
from keras.layers import Dense, Activation, core
from keras.optimizers import SGD
import keras.regularizers as regl

from sklearn.preprocessing import OneHotEncoder, MinMaxScaler

import numpy as np

wineArray = np.loadtxt('wine.data',dtype=float,delimiter=',')
wineLabels = wineArray[:,0].reshape(-1,1)
winePred = MinMaxScaler().fit_transform(wineArray[:,1:])


OHE = OneHotEncoder()
wineLabels = OHE.fit_transform(wineLabels).toarray()
#wineLabels[wineLabels==0] = -1
```

## Training

```python
model = Sequential()
model.add(Dense(20, input_dim=13, activation='tanh' ))
core.Dropout(.1)
model.add(Dense(25, activation='tanh', W_regularizer=regl.l1(.01))) # >1 reguralizers doesn't converge with 250 epochs
core.Dropout(.1)
model.add(Dense(3, activation='softmax'))

sgd = SGD(lr=0.1, momentum=0.5, nesterov=True)
model.compile(loss='mse', optimizer=sgd)
# model.compile(loss='mae', optimizer=sgd)
# model.compile(loss='msle', optimizer=sgd)
# model.compile(loss='categorical_crossentropy', optimizer=sgd)

hist = model.fit(winePred, wineLabels, batch_size=50, nb_epoch=250, verbose=0, show_accuracy=True, validation_split=.1, shuffle=True).history
# Lists the keys of the history object
for x in hist:
    print(x)
print(hist)
# model.fit(winePred, wineLabels, batch_size=50, nb_epoch=1, show_accuracy = True, verbose=1)
```

## Prediction
Predicts the output values of the 100-150th lines of the trained input data

```python
model.predict(winePred[100:150,:], verbose=1) #.reshape((1,-1)))
```
